```python
noise_case = 1200 # 2nd plot

base_path = os.getcwd()

_, samps_LV= pos_samples_modified(filename=base_path + '/../LV/post_samples/post_samples_interp_L1V1_{}.hdf'.format(noise_case))
                        
_, samps_LVA = pos_samples_modified(filename=base_path + '/../LVA/post_samples/post_samples_interp_L1V1A0_{}.hdf'.format(noise_case))
```

```python
methods = ['L1V1', 'L1V1A0']

labels = [r'$\mathcal{M} \,\, (M_{\odot})$', r'$\chi_{eff}$', r'$\iota \,\,(rad)$', \
                              r'$\alpha \,\,(rad)$', r'$\delta \,\,(rad)$', r'$d_L \, \,(Mpc)$', r'$z$']

titles = [r'$\mathcal{M}$', r'$\chi_{eff}$', r'$\iota$', r'$\alpha$', r'$\delta$', r'$d_L$', r'$z$']

fmt_arr = ['.5f', '.5f', '.3f', '.3f', '.3f', '.1f', '.4f']

#-- samples list --

samps_comb = [samps_LV, samps_LVA]


range_vals = []

for i in range(samps_LV.shape[1]):
    
    if(i==3):
        
        range_vals.append((3.8, 4.8))
        
    elif(i==4):
        
        #range_vals.append((-0.5, 0.8))
        range_vals.append((samps_LV[:,i].min(), samps_LV[:,i].max()))
    else:
        
        range_vals.append((samps_LV[:,i].min(), samps_LV[:,i].max()))
        # print(range_vals)

lw = 1.5

measure = 'median'

bins = 20

fontsize = 18

tlabels_1 = title_formats(samps_LV, labels, titles, fmt_arr, measure=measure, bins=bins, flag='first')
tlabels_2 = title_formats(samps_LVA, labels, titles, fmt_arr, measure=measure, bins=bins, flag='second')

tlabels = [tlabels_1, tlabels_2]

kwargs = dict(labels=labels, titles=titles, smooth=0.9, levels=[0.90], truths=truths, tlabels=tlabels, bins=bins, labelpad=0.07, label_kwargs=dict(fontsize=fontsize), contour_kwargs=dict(linewidths=lw), \
              hist_kwargs=dict(lw=lw, density=True), title_kwargs=dict(fontsize=fontsize), plot_density=False, \
              methods=methods, plot_datapoints=False)

filename = 'n{}_GW190425_corner_plot.png'.format(noise_case)
#fig, axes = plot_corner_2datasets(samples=samps_comb, colors=['crimson', 'blue'], linestyle=['-', '-'], \
#                                  measure=measure, filename=filename, save=False, **kwargs);

fig = plot_corner_2datasets(samples=samps_comb, colors=['crimson', 'dodgerblue'], linestyle=['-', '-'], \
                            measure=measure, filename=filename, save=True, **kwargs);
```
