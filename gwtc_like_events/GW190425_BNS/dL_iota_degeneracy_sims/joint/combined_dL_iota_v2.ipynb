```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import corner
import os
import seaborn as sns
import pandas as pd
```

#### Realizations with baseline of injected mchirp being within 90% CI in both LV and LVA (28)

```python
LV_address = os.getcwd() + '/../LV/post_samples/'
LVA_address = os.getcwd() + '/../LVA/post_samples/'

sims = np.arange(200, 6200, 100)

mchirp_true = 1.4847507357166811
dL_LV = []
iota_LV = []
idx_LV = []

for i in sims:
    
    file = h5py.File(LV_address + 'post_samples_interp_L1V1_{}.hdf'.format(i), 'r')
    temp = np.array(file['Mc'])
    
    if(mchirp_true >= np.quantile(temp, q=[0.05]) and mchirp_true <= np.quantile(temp, q=[0.95])):
        
        dL_LV.append(list(np.array(file['dL'])))
        iota_LV.append(list(np.array(file['iota'])))
        idx_LV.append(i)
        
    file.close()
    
dL_LVA = []
iota_LVA = []
idx_LVA = []

for i in idx_LV:
    
    file = h5py.File(LVA_address + 'post_samples_interp_L1V1A0_{}.hdf'.format(i), 'r')
    temp = np.array(file['Mc'])
    if(mchirp_true >= np.quantile(temp, q=[0.05]) and mchirp_true <= np.quantile(temp, q=[0.95])):
        
        dL_LVA.append(list(np.array(file['dL'])))
        iota_LVA.append(list(np.array(file['iota'])))
        idx_LVA.append(i)
        
    file.close()

idx_com = np.intersect1d(idx_LV, idx_LVA)
dL_LV_fin = []
dL_LVA_fin = []
iota_LV_fin = []
iota_LVA_fin = []

for i in idx_com:
    
    file1 = h5py.File(LV_address + 'post_samples_interp_L1V1_{}.hdf'.format(i), 'r')
    file2 = h5py.File(LVA_address + 'post_samples_interp_L1V1A0_{}.hdf'.format(i), 'r')
        
    dL_LV_fin.append(list(np.array(file1['dL'])))
    iota_LV_fin.append(list(np.array(file1['iota'])))
    
    dL_LVA_fin.append(list(np.array(file2['dL'])))
    iota_LVA_fin.append(list(np.array(file2['iota'])))
        
file1.close()
file2.close()

iota_LV_fin = np.array(np.array(iota_LV_fin).flatten()[0])
dL_LV_fin = np.array(np.array(dL_LV_fin).flatten()[0])

iota_LVA_fin = np.array(np.array(iota_LVA_fin).flatten()[0])
dL_LVA_fin = np.array(np.array(dL_LVA_fin).flatten()[0])

LV_str = []
LVA_str = []

for i in range(len(iota_LV_fin)):
    
    LV_str.append('L1V1')
    
for i in range(len(iota_LVA_fin)):
    
    LVA_str.append('L1V1A1')
    
data1 = {r'$\iota$': iota_LV_fin, r'$d_L$': dL_LV_fin, 'Network': LV_str}  
df1 = pd.DataFrame(data1)

data2 = {r'$\iota$': iota_LVA_fin, r'$d_L$': dL_LVA_fin, 'Network': LVA_str}  
df2 = pd.DataFrame(data2)
    
df_merged = pd.concat([df1, df2])

height=6
ratio=4
fontsize = 12
lw = 1.5
 
g = sns.jointplot(data=df_merged, x=r"$\iota$", y=r"$d_L$", \
              marginal_kws={"lw": lw, 'fill':True}, height=height, ratio=ratio, hue="Network", \
              palette={'L1V1': 'crimson', 'L1V1A1': 'dodgerblue'}, marker='.', s=5) #,joint_kws={'marker': '.', 'size': 0.1}) #, kind='kde', n_levels=1)

g.ax_joint.tick_params(labelsize=fontsize)
g.ax_joint.set_xlabel(r'$\iota$ [rad]', fontsize=fontsize)
g.ax_joint.set_ylabel(r'$d_L$ [Mpc]', fontsize=fontsize)
g.ax_joint.set_xticks(np.arange(0, np.pi+0.01, np.pi/4))
labels = ['$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$']
g.ax_joint.set_xticklabels(labels)
g.ax_joint.scatter(0.53, 171, marker='*', c='k', s=100, alpha=1, label='Injection')
g.ax_joint.legend(loc=4, fontsize=fontsize)
g.ax_joint.set_xlim(0, np.pi)
g.ax_joint.axvline(x=0.53, ls='--', c='grey', lw=0.8, zorder=2)
g.ax_joint.axhline(y=171, ls='--', c='grey', lw=0.8, zorder=2)
g.savefig('dL_iota_sims_GW190425_28rels_A1.png', dpi=200, bbox_inches='tight')
```
