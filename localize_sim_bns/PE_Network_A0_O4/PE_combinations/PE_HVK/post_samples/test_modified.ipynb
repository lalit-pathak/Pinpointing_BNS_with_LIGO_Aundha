```python
import h5py
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
with h5py.File('quantities_of_interest.hdf', 'a') as f:
    
    print(f.keys())
```

```python
with h5py.File('quantities_of_interest.hdf', 'a') as f:
    
    delta_omega = np.array(f['L1H1V1K1']['delta_omega'])
```

```python
def ecdf(a):
    x, counts = np.unique(a, return_counts=True)
    cusum = np.cumsum(counts)
    return x, cusum / cusum[-1]

vals, frac = ecdf(delta_omega)


fig, ax = plt.subplots(figsize=(8, 5))

fontsize = 12

ax.plot(vals, frac, lw=2)
ax.set_xlabel(r'$\Delta \, \Omega \,[deg^2]$ ', fontsize=fontsize)
ax.set_ylabel(r'Fraction $< \,\Delta \, \Omega$', fontsize=fontsize)
ax.tick_params(labelsize=fontsize)
ax.set_xscale("log")
ax.grid()
fig.show()
```

```python
kde = gaussian_kde(np.sort(delta_omega))
plt.plot(np.sort(delta_omega), kde(np.sort(delta_omega)))
```
