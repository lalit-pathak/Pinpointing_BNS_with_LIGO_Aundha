```python
#-- import modules -

import numpy as np
import matplotlib.pyplot as plt
import h5py
import pandas as pd
from itertools import combinations
import os

from pycbc.detector import add_detector_on_earth
from pycbc.waveform import get_waveform_filter_length_in_time as duration
from pycbc.psd.analytical import aLIGOAPlusDesignSensitivityT1800042, AdVDesignSensitivityP1200087, KAGRADesignSensitivityT1600593, KAGRA, AdvVirgo
from pycbc.psd.analytical import aLIGODesignSensitivityP1200087, aLIGOZeroDetHighPower
from pycbc.cosmology import cosmological_quantity_from_redshift
from pycbc.cosmology import redshift
from pycbc.cosmology import distance_from_comoving_volume, redshift_from_comoving_volume
from pycbc.waveform.generator import FDomainCBCGenerator, FDomainDetFrameGenerator
from pycbc.types.frequencyseries import load_frequencyseries
from pycbc.filter import sigma
from pycbc.psd.estimate import interpolate
from pycbc.types import FrequencySeries
from pycbc.noise.gaussian import frequency_noise_from_psd
from pycbc.filter.matchedfilter import matched_filter
from pycbc.waveform import get_fd_waveform
from pycbc.noise.gaussian import frequency_noise_from_psd
```

```python
#-- adding LIGO India to PyCBC detectors --

add_detector_on_earth(name='A0', longitude=1.34444215058, latitude=0.34231676739,\
                      yangle=4.23039066080, xangle=5.80120119264, height=440) 

```

```python
#-- Choosing O4 PSD for LIGO India --

base_path = os.getcwd() + '/../../../../psds/'

ifos = ['L1', 'H1', 'V1', 'K1', 'A0']

psds = {}

for ifo in ifos:
    
    if(ifo=='V1'):
        
        psds[ifo] = load_frequencyseries(base_path + 'PSD_Virgo_10Hz_to_2048Hz_nonzero.txt')
        
    elif(ifo=='K1'):
        
        psds[ifo] = load_frequencyseries(base_path + 'PSD_KAGRA_10Hz_to_2048Hz_nonzero.txt')
        
    elif(ifo=='A0'):
        
        psds[ifo] = load_frequencyseries(base_path + 'PSD_O4_10Hz_to_2048Hz_nonzero.txt')
        
    else:
        
        psds[ifo] = load_frequencyseries(base_path + 'PSD_O5_10Hz_to_2048Hz_nonzero.txt')
```

```python
#-- intrinsic parameters --

m1 = 1.387
m2 = 1.326
s1z = 0.0001294116725339975
s2z = 3.5461352589076355e-05

#-- Waveform Approximants used --

inj_approx = 'IMRPhenomD'
```

```python
#-- setting up the required params for PSD generation --

fLow, fHigh = 10, 1600    # Hz
segLen = 1500             # secs (duration of signal) (segLen > 1178.3 secs, and hence the template(1178.3 secs long) will be able to cover the DATA(1500 secs long) which is to be generated)
fSamp = 4096              # Hz   (Sampling Frequency)

delta_f = 1/segLen        # Hz
fLen = int(segLen*fSamp/2 + 1)      # No of Frequency Samples
```

```python
#-- extrinsic parameters --

iota = np.pi/6
pol = 0

address = os.getcwd() + '/../../../PE_Network_A0_O4/detection_criteria_bns/injections_L1H1V1K1A0_O4_SNR_20_to_25.txt'
ra, dec, dL, z, tc = np.loadtxt(address)[69,:5]  # count = 71, and in file index = 70

ra, dec, dL, z, tc
```

```python
#-- generating detector-frame injection --

static_params_gen = {'approximant': inj_approx, 'f_lower': fLow, 'mass1': m1*(1+z), 'mass2': m2*(1+z),\
                     'spin1z': s1z, 'spin2z': s2z , 'polarization': pol, 'inclination': iota, 'tc': tc,\
                     'distance': dL, 'ra': ra, 'dec': dec}
            
#-- NOTE: Here signal is created in five_dets (including A0) but SNRs will be calculated for four_dets --
generator = FDomainDetFrameGenerator(FDomainCBCGenerator, epoch=static_params_gen['tc'] - 1300 , detectors=ifos,\
                                     delta_f = 1/segLen, **static_params_gen)

signal = generator.generate() 
```

```python
hp, hc = get_fd_waveform(approximant='TaylorF2', mass1=m1*(1+z), mass2=m2*(1+z),\
                         spin1z=s1z, spin2z=s2z, delta_f = 1/segLen, f_lower = fLow)

```

```python
noise_tilde = {}

noise_seed = 1
count = 70

i = 0
j = 10 


for ifo in ifos:
    
    if(ifo=='A0'):
        noise_tilde[ifo] = frequency_noise_from_psd(psd=psds[ifo], seed=noise_seed+j)
            
    else:    
        noise_tilde[ifo] = frequency_noise_from_psd(psd=psds[ifo], seed=noise_seed+i)
        i +=1
```

```python
#-- calculating optimal SNR --

htilde = {}
sim_data = {}
snrs = {}

net_optimal_snr = 0

for ifo in ifos:  

    htilde[ifo] = signal[ifo]                      #-- Simulated Detector-Frame GW signal
    htilde[ifo].resize(len(psds[ifo]))
    
    hp.resize(len(htilde[ifo]))

    #-- calculating optimal SNR using matched-filter function in PyCBC  
    
    snrs[ifo] = matched_filter(hp, htilde[ifo], psd=psds[ifo], low_frequency_cutoff=fLow, high_frequency_cutoff=fHigh)
    
  #  snrs[ifo] = snrs[ifo].time_slice(start=tc - 0.07, end=tc + 0.07)
        
 #   snrs[ifo] = abs(snrs[ifo]).max()
        
 #   net_snr += (snr_temp[ifo])**2
    
    net_optimal_snr += (abs(snrs[ifo]).max())**2

 #   net_optimal_snr += (abs(snrs[ifo])**2).max()

net_optimal_snr = np.sqrt(net_optimal_snr)
```

```python
net_optimal_snr
```

```python
plt.figure(figsize=(15, 6))
for ifo in ifos:
    
    plt.plot(snrs[ifo].sample_times, abs(snrs[ifo]), label='%s : %.1f'%(ifo, abs(snrs[ifo]).max()))
    plt.xlim(tc - 0.50, tc + 0.50)

plt.title('Network SNR : %.1f'%net_optimal_snr)
plt.xlabel('GPS Times(s)')
#plt.axvline(tc, ls='--', c='k')
plt.ylabel('SNR')
plt.legend()
#plt.savefig("SNR_4Det_network_test2.jpg", dpi=200, bbox_inches='tight')
plt.show()
```

```python
#-- calculating matched-filter SNR --

htilde = {}
sim_data = {}
snrs = {}
snr_val = {}

net_snr = 0

for ifo in ifos:  

    htilde[ifo] = signal[ifo]                      #-- Simulated Detector-Frame GW signal
    htilde[ifo].resize(len(psds[ifo]))
    
    sim_data[ifo] = htilde[ifo] + noise_tilde[ifo]
    
    hp.resize(len(htilde[ifo]))

    #-- calculating optimal SNR using matched-filter function in PyCBC  
    
    snrs[ifo] = matched_filter(hp, sim_data[ifo], psd=psds[ifo], low_frequency_cutoff=fLow, high_frequency_cutoff=fHigh)
    
    
#     snrs[ifo].crop(500, 500)
#     snrs[ifo].max()
    
#     net_snr += (abs(snrs[ifo])**2)

    #-- around 70 ms from trigger time --
    snrs[ifo] = snrs[ifo].time_slice(start=tc - 0.07, end=tc + 0.07)
        
    snr_val[ifo] = abs(snrs[ifo]).max()
        
    net_snr += (snr_val[ifo])**2

net_snr = np.sqrt(net_snr)
```

```python
plt.figure(figsize=(15, 4))
fontsize=14

for ifo in ifos:
    
    plt.plot(snrs[ifo].sample_times, abs(snrs[ifo]), label='%s : %.1f'%(ifo, snr_val[ifo]))
   # plt.xlim(tc - 0.20, tc + 0.20)

plt.title('Network SNR : %.1f'%net_snr, fontsize=fontsize)
plt.xlabel('GPS Times(s)', fontsize=fontsize)
#plt.axvline(tc, ls='--', c='k')
plt.ylabel('SNR', fontsize=fontsize)
plt.legend()
#plt.savefig("SNR_4Det_network_test2.jpg", dpi=200, bbox_inches='tight')
plt.show()
```
