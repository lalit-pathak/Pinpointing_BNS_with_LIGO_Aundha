## $\Rightarrow$ For L1H1V1K1 Network:

````python
network_4 = ['L1', 'H1', 'V1', 'K1']

duty_rate_4 = {'L1': 0.8, 'H1': 0.8, 'V1': 0.8, 'K1': 0.8}

det_num_allowed_4 = [3, 4]

cp_4, tp_4 = combination_prob_calculator(network=network_4, duty_rate=duty_rate_4, det_num_allowed=det_num_allowed_4)

````

### $\Rightarrow$ Selecting events for different combinations:

````python
event_4Det_seed = 0

event_4Det_idx = {}

seed_4Det_list = []

for key, val in network_events_4['L1H1V1K1'].items():
    
    np.random.seed(event_4Det_seed)
    seed_4Det_list.append(event_4Det_seed)
    event_4Det_idx[key] = np.random.choice(a=np.arange(0, total_events, 1), size=int(val), replace=False)
    
    event_4Det_seed += 1
````

## $\Rightarrow$ For L1H1V1K1A0 network:

````python
network_5 = ['L1', 'H1', 'V1', 'K1', 'A0']

A0_duty_rate = 0.5

duty_rate_5 = {'L1': 0.8, 'H1': 0.8, 'V1':0.8, 'K1':0.8, 'A0': A0_duty_rate}

det_num_allowed_5 = [3, 4, 5]

cp_5, tp_5 = combination_prob_calculator(network=network_5, duty_rate=duty_rate_5, det_num_allowed=det_num_allowed_5)
````

#### Note: The below function ```network_events_calculator()``` would require ```net_comb_prob_list``` with both the network combination probabilities for L1H1V1K1A0 case, since it depends on the highest probability based on L1H1V1K1

````python
#-- to be given as a list --

net_comb_prob_list = [cp_4, cp_5]

network_events = network_events_calculator(net_comb_prob_list, total_events=total_events)

network_events
````

#### Selecting the events per detector combination (network) based on their observing probabilities:

````python
#- for LHVKA (5 Detector Network Case) -

event_5Det_seed = 5

event_5Det_idx = {}

seed_5Det_list = []

for key, val in network_events['L1H1V1K1A0'].items():

    if(key=='L1H1V1'):

        np.random.seed(seed = seed_4Det_list[0])
        event_5Det_idx[key] = np.random.choice(event_4Det_idx[key], size=int(np.round(val+0.1)), replace=False)
        #print(event_5Det_idx[key])
        
    elif(key=='L1H1K1'):

        np.random.seed(seed = seed_4Det_list[1])
        event_5Det_idx[key] = np.random.choice(event_4Det_idx[key], size=int(np.round(val+0.1)), replace=False)

    elif(key=='L1V1K1'):

        np.random.seed(seed = seed_4Det_list[2])
        event_5Det_idx[key] = np.random.choice(event_4Det_idx[key], size=int(np.round(val+0.1)), replace=False)

    elif(key=='H1V1K1'):

        np.random.seed(seed = seed_4Det_list[3])
        event_5Det_idx[key] = np.random.choice(event_4Det_idx[key], size=int(np.round(val+0.1)), replace=False)

    elif(key=='L1H1V1K1'):

        np.random.seed(seed = seed_4Det_list[4])
        event_5Det_idx[key] = np.random.choice(event_4Det_idx[key], size=int(np.round(val+0.1)), replace=False)

    else:

        np.random.seed(seed=event_5Det_seed)
        seed_5Det_list.append(event_5Det_seed)
        event_5Det_idx[key] = np.random.choice(a=np.arange(0, total_events, 1), size=int(np.round(val+0.1)), replace=False)

        event_5Det_seed += 1
````

````python
#-- delta_omega 90% for LHVK combinations (As observed) --
    
d_omega_LHVK = {}

for key, vals in event_4Det_idx.items():

    d_omega_LHVK[key] = data_O4[key]['delta_omega'][event_4Det_idx[key]]   #-- 3rd bracket is for valid indices

#- weighted delta_omega array -

#d_omega_W4 = np.concatenate((d_omega_LHVK['L1H1V1'], d_omega_LHVK['L1H1K1'], d_omega_LHVK['L1V1K1'],\
#                            d_omega_LHVK['H1V1K1'], d_omega_LHVK['L1H1V1K1']))



#- weighted delta_omega array --
d_omega_W4 = np.empty(0)

for key, vals in d_omega_LHVK.items():
    
    d_omega_W4 = np.concatenate((d_omega_W4, d_omega_LHVK[key]))
````

````python
#-- delta_omega 90% for LHVKA combinations (As observed with A0 at O4) --

d_omega_LHVKA_O4 = {}

for key, vals in event_5Det_idx.items():

    d_omega_LHVKA_O4[key] = data_O4[key]['delta_omega'][event_5Det_idx[key]]   #-- 3rd bracket is for valid indices

#- weighted delta_omega array --

d_omega_W5_O4 = np.empty(0)

for key, vals in d_omega_LHVKA_O4.items():
    
    d_omega_W5_O4 = np.concatenate((d_omega_W5_O4, d_omega_LHVKA_O4[key]))
````

### $\Rightarrow$ Plotting the CDF for $\Delta \Omega_{90\%}$

````python
num_logspace_hist = 1000

bins = np.logspace(np.log10(1e-1), np.log10(1e+3), num=num_logspace_hist)
````

````python
plt.plot(bins)
plt.show()
````

````python
#-- LHVK Network --

cdf_4Det = {}
val_4Det = {}

for key, vals in d_omega_LHVK.items():

    counts, values = np.histogram(vals, bins=bins)

    cdf_temp = np.cumsum(counts)/np.cumsum(counts)[-1]

    val_4Det[key], cdf_4Det[key] = values, cdf_temp


#-- weighted for 4 detector network --

f_4Det, v_4Det = np.histogram(d_omega_W4, bins=bins)        #-- See Events with A0 at O4 (first code snippet)

# new cdf
cdf_w4 = np.cumsum(f_4Det)/np.cumsum(f_4Det)[-1]

# new values
val_w4 = v_4Det 
````

````python
#-- LHVKA Network with LIGO Aundha at O4--
cdf_5Det_O4 = {}
val_5Det_O4 = {}

for key, vals in d_omega_LHVKA_O4.items():

    counts, values = np.histogram(vals, bins=bins)

    cdf_temp = np.cumsum(counts)/np.cumsum(counts)[-1]

    val_5Det_O4[key], cdf_5Det_O4[key] = values, cdf_temp

#-- Weighted for 5 detector network with LIGO Aundha at O4--

f_O4, v_O4 = np.histogram(d_omega_W5_O4, bins=bins)

# new cdf
cdf_w5_O4 = np.cumsum(f_O4)/np.cumsum(f_O4)[-1]

# new values
val_w5_O4 = v_O4
````

````python
#-- LHVKA Network with LIGO Aundha at at O5--
cdf_5Det_O5 = {}
val_5Det_O5 = {}

for key, vals in d_omega_LHVKA_O5.items():

    counts, values = np.histogram(vals, bins=bins)

    cdf_temp = np.cumsum(counts)/np.cumsum(counts)[-1]

    val_5Det_O5[key], cdf_5Det_O5[key] = values, cdf_temp

#-- Weighted for 5 detector network with LIGO Aundha at O5--

f_O5, v_O5 = np.histogram(d_omega_W5_O5, bins=bins)

# new cdf
cdf_w5_O5 = np.cumsum(f_O5)/np.cumsum(f_O5)[-1]

# new values
val_w5_O5 = v_O5
````
